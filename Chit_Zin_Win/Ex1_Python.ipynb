# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
num_sec = int(input("Enter the number of seconds: "))
minutes = num_sec // 60
seconds = num_sec % 60
print("{} seconds is equal to {} minutes and {} seconds.".format(num_sec,minutes,seconds))
```
